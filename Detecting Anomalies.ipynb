### Reference columns
###### MaterialID: represents the production lot
###### StepID: represents the step of the production
###### duration_ms: represents the time elapsed (normalized) from the first time sample (in mandatory step) which has duration_ms 0, the last time sample (in mandatory step) has duration_ms equal to 1. Time samples in optional step have duration_ms greater than 1 or lesser than 0.
###### Target: is a boolean value that represent if the MaterialID is abnormal or not, 1 for abnormal MaterialID, 0 for normal MaterialID.
###### is_test: is a boolean value that represent if the time sample has been used as training or test in our work. 1 means that the time sample has been used as test, 0 means that the time sample has been used for the training phase.
https://github.com/STMicroelectronics/ST-AWFD/blob/main/README.md

## Wafer 1 - Train/Test

```python
from sklearn.metrics import accuracy_score
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
target_col = "target"
drop_cols = ["target", "is_test", "MaterialID", "StepID"]

X = df1.drop(columns=drop_cols)
y = df1[target_col]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

## Wafer 1 - XGBoost

```python
# Apply SMOTE
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve

smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Calculate scale_pos_weight (ratio of majority/minority in original training set)
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

# Train XGBoost with both resampled data + class weight
model = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss",
    scale_pos_weight=scale_pos_weight
)

model.fit(X_train_resampled, y_train_resampled)

# Evaluate on test set
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
y_scores = model.predict_proba(X_test)[:, 1]

# Compute precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

# Plot the curve
plt.figure(figsize=(8,6))
plt.plot(thresholds, precision[:-1], label="Precision")
plt.plot(thresholds, recall[:-1], label="Recall")
plt.xlabel("Decision Threshold")
plt.ylabel("Score")
plt.title("Precision-Recall vs Threshold")
plt.legend()
plt.grid(True)
plt.show()

# Example: choose a threshold (say 0.7) to balance precision and recall
chosen_threshold = 0.7
y_pred_custom = (y_scores >= chosen_threshold).astype(int)

# Evaluate with the custom threshold
print(classification_report(y_test, y_pred_custom))
```

## Wafer 1 - Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
# Apply SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)


# initialize model

iso = IsolationForest(
    n_estimators=100,
    contamination=0.01,
    max_samples=256,   # default, efficient for large datasets
    random_state=42
)

iso.fit(X_train_resampled)
```

```python
# Isolation Forest predicts: -1 (anomaly), +1 (normal)
y_pred_test = iso.predict(X_test)

# Map to your convention: 1 = anomaly, 0 = normal
y_pred_test = np.where(y_pred_test == -1, 1, 0)

# Now compare properly
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```

```python
from sklearn.metrics import precision_recall_curve

scores = iso.decision_function(X_test)
prec, rec, thresh = precision_recall_curve(y_test, -scores)  # flip sign for anomalies
prec, rec, thresh

plt.plot(rec, prec)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve for Isolation Forest")
plt.show()
```

## Wafer 1- Autoencoder

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# IMPORTANT: Train autoencoder only on normal class (y == 0)
X_train_normal = X_train[y_train == 0]

# Define autoencoder architecture
input_dim = X_train.shape[1]  # number of features
encoding_dim = 16             # size of latent space (tuneable)

input_layer = layers.Input(shape=(input_dim,))
encoded = layers.Dense(encoding_dim, activation='relu')(input_layer)
decoded = layers.Dense(input_dim, activation='sigmoid')(encoded)

autoencoder = models.Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# Train autoencoder on normal data only
autoencoder.fit(X_train_normal, X_train_normal,
                epochs=50,
                batch_size=128,
                shuffle=True,
                validation_split=0.1,
                verbose=2)

# Reconstruction error on test set
X_test_pred = autoencoder.predict(X_test)
reconstruction_error = np.mean(np.square(X_test - X_test_pred), axis=1)

# Choose threshold (e.g., 95th percentile of normal errors)
threshold = np.percentile(reconstruction_error[y_test == 0], 95)

# Predict anomalies: error above threshold = anomaly
y_pred = (reconstruction_error > threshold).astype(int)

# Evaluate
print(classification_report(y_test, y_pred))
```

```python
plt.hist(reconstruction_error[y_test==0], bins=50, alpha=0.7, label="Normal")
plt.hist(reconstruction_error[y_test==1], bins=50, alpha=0.7, label="Anomaly")
plt.xlabel("Reconstruction error")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, reconstruction_error)
f1_scores = 2*precision*recall/(precision+recall+1e-8)
best_threshold = thresholds[np.argmax(f1_scores)]
print("Best threshold:", best_threshold)
```
